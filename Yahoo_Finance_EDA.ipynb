# Imports

```python
import datetime as dt
import pandas as pd
from pandas_datareader import data as pdr
import yfinance as yf
import matplotlib.pyplot as plt
yf.pdr_override()
```

# Date Range

```python
end = dt.datetime.now()
start = end - dt.timedelta(days = 1000)

start,end
```

# Select stocks/tickers to analyze

```python
stocks = ['JPM', 'BAC', 'WFC', 'COF']
# stocks = ['AAPL.AX', 'COF.AX']
```

# Pandas datareader

```python
df = pdr.get_data_yahoo(stocks, start, end)

df.head()
```

```python
df.columns
```

# Exploration

```python
close = df.Close
vol = df.Volume
```

```python
close.describe()
```

# Plotting

```python
close.plot(figsize = (8,5))
plt.title("Daily stock prices")
plt.ylabel("USD")
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")
plt.show()
```
